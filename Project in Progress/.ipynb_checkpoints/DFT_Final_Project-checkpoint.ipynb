### Mark's Questions:-

# Q1:
    Has the rooms availability in the homeless shelters been affected by the covid cases in the county of San Francisco?

We'll run Independent Chi-Square to determine whether "cumulative_homeless_cases" variable is influencing "rooms_available" variable which will answer the availablity question. 

### 4a: Recoding variables into categorical variables

Recoding 'cumulative_homeless_cases' into a categorical variable

```python
def homeless_cases (series): 
    if series <= 165:
        return 0
    if series > 165: 
        return 1
```

```python
covid19_homeless_impact_Final['cumulative_homeless_cases_R'] = covid19_homeless_impact_Final['cumulative_homeless_cases'].apply(homeless_cases)
```

```python
sns.distplot(covid19_homeless_impact_Final['cumulative_homeless_cases_R'])
```

Recoding 'rooms_available' into a categorical variable

```python
def rooms_ava (series): 
    if series <= 866:
        return 0
    if series > 866: 
        return 1
```

```python
covid19_homeless_impact_Final['rooms_available_R'] = covid19_homeless_impact_Final['rooms_available'].apply(rooms_ava)
```

```python
sns.distplot(covid19_homeless_impact_Final['rooms_available_R'])
```

### 4b: Create a Contingency Table

```python
covid19_homeless_impact_Final_CrossTab = pd.crosstab(covid19_homeless_impact_Final['cumulative_homeless_cases_R'], covid19_homeless_impact_Final['rooms_available_R'])
```

```python
covid19_homeless_impact_Final_CrossTab
```

### 4c: Running the Independent Chi-Square

```python
scipy.stats.chi2_contingency(covid19_homeless_impact_Final_CrossTab)
```

We conclude that as the homeless cases increased, the rooms availability decreased in the county of San Francisco especially in the later months of June and July.

# Testing assumptions:-

### 1- Testing for Normality

```python
sns.distplot(covid19_homeless_impact_Final['homless_population_by_percentage'])
```

```python
sns.distplot(covid19_homeless_impact_Final['general_public_population_by_percentage'])
```

Both variables are normally distributed

### 2- Testing for Homoscedasticity

We will create the Basic Model

```python
x = covid19_homeless_impact_Final['homless_population_by_percentage']
y = covid19_homeless_impact_Final['general_public_population_by_percentage']
```

```python
SFmodel = sm.OLS(y,x).fit()
```

calculating out the residuals

```python
pred_val = SFmodel.fittedvalues.copy()
true_val = covid19_homeless_impact_Final['general_public_population_by_percentage'].values.copy()
residual = true_val - pred_val
```

 Graphing residuals to observe homoscedasticity

```python
fig, ax = plt.subplots(figsize=(6, 2.5))
_ = ax.scatter(residual, pred_val)
```

The graph shows some residuals and we unsure if we could have a straight line run through all the values. So we will check the graph using the Breusch Pagan test from the sms package.

```python
sms.diagnostic.het_breuschpagan(residual, covid19_homeless_impact_Final[['homless_population_by_percentage']])
```

The output shows the F value is 70, this is a big number which increases chances of being statistically significant and the pvalue of 6.7e-13 < .05 is statistically significant , so we have violated the assumption of homoscedasticity and the data in fact is heteroscedastic. We will try again using the Harvey Collier test.

```python
sms.linear_harvey_collier(SFmodel)
```

This test also reveals a pvalue < .05, which violates assumption of homoscedasticity. So we will transform our variables so they become homoscedastic. We will apply the Box-cox transformation.

The pvalue is still significant , in this case, we will remove the outliers in order to continue with the Analysis

### Screen for outliers

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(SFmodel, alpha = .05, ax = ax, criterion="cooks")
```

Remove the outliers row:85-89

```python
SF2 = covid19_homeless_impact_Final[:60]
SF2
```

```python
x = SF2['homless_population_by_percentage']
y = SF2['general_public_population_by_percentage']
SF2model = sm.OLS(y,x).fit()
pred_val = SF2model.fittedvalues.copy()
true_val = SF2['general_public_population_by_percentage'].values.copy()
residual = true_val - pred_val
fig, ax = plt.subplots(figsize=(6, 2.5))
_ = ax.scatter(residual, pred_val)
```

```python
sms.linear_harvey_collier(SF2model)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(SF2model, alpha = .05, ax = ax, criterion="cooks")
```

We can observe that we have 4 outliers labeled in the graph. Lets look a a table as well to double check.

We do not have to Test for Multicollinearity because we have only one independent variable.

```python
infl = SF2model.get_influence()
print(infl.summary_frame())
```

For the student_resid, we see that we have a value that over 2.5 in row 89, this means that we have an outlier problem.

```python
SF2model.summary()
```

The overall model is statistically significant because pvalue <.05. the homeless cases have a significant effect on the number of general public cases. When we look at R- squared and Adj. R-squared it shows that this model accounts for 92%of the variance in explaining general public cases.
